```
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.preprocessing import LabelEncoder, StandardScaler
data={
    'size':['L','M','S','XS','XL','M','S','L','XXL','L','M'],
    'color':['green','red','yellow','black','grey','blue','green','red','yellow','black','grey'],
    'gender':['female','female','male','male','female','male','male','female','female','female','male'],
    'price':[149.0,79.0,152.0,123.0,321.0,32.0,23.0,49.0,79.0,152.0,84.0],
    'weight':[380,410,320,420,415,123,412,321,451,500,465],
    'bought':['no','yes','no','yes','no','yes','no','yes','no','yes','yes']
    }
df_raw=pd.DataFrame(data=data)
df=df_raw.copy()
for col in['size','color','gender','bought']:
  df[col]=df[col].astype('category') #nie do konca zrozumialem jak mam to zmienic na int i float zeby to zadzialalo
df.describe(include=['category']).T
le=LabelEncoder()
le.fit_transform(df['bought'])
pd.get_dummies(data=df,drop_first=True)
scaler=StandardScaler()
df[['price','weight']]=scaler.fit_transform(df[['price','weight']])
```
